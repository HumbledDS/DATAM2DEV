## Premier pas avec Python et les données

### Étape 1 : Importer les bibliothèques essentielles

Nous allons utiliser les bibliothèques suivantes :
- **pandas** : manipulation et analyse de données
- **numpy** : calculs numériques
- **matplotlib** : visualisation basique

```python
# Imports des bibliothèques principales
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import warnings

# Paramètres de configuration pour les visualisations
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
warnings.filterwarnings('ignore')

print('Toutes les bibliothèques sont importées avec succès!')
```

### Étape 2 : Charger et explorer le jeu de données Iris

Le dataset Iris est l'un des plus célèbres en machine learning. Il contient des mesures de fleurs iris avec 4 caractéristiques (features) et 3 classes.

```python
# Charger le dataset Iris depuis scikit-learn
iris = datasets.load_iris()

# Créer un DataFrame pandas pour une meilleure exploration
df_iris = pd.DataFrame(
    data=iris.data,
    columns=iris.feature_names
)

# Ajouter la colonne cible (target - les espèces)
df_iris['species'] = iris.target_names[iris.target]

print('Forme du dataset:', df_iris.shape)
print('\nPremières lignes du dataset:')
print(df_iris.head())
```

```python
# Informations détaillées sur le dataset
print('Informations sur les colonnes:')
print(df_iris.info())

print('\n' + '='*50)
print('Statistiques descriptives:')
print(df_iris.describe())

print('\n' + '='*50)
print('Nombre d\'exemples par espèce:')
print(df_iris['species'].value_counts())
```

### Étape 3 : Visualisation des données - Scatter plot

Créons un graphique en nuage de points pour visualiser la relation entre la longueur et la largeur du sépale, en colorant les points selon l'espèce.

```python
# Créer une figure pour la visualisation
plt.figure(figsize=(10, 6))

# Créer un scatter plot pour chaque espèce
for species in df_iris['species'].unique():
    data_species = df_iris[df_iris['species'] == species]
    plt.scatter(
        data_species['sepal length (cm)'],
        data_species['sepal width (cm)'],
        label=species,
        s=100,
        alpha=0.7
    )

plt.xlabel('Longueur du sépale (cm)', fontsize=12)
plt.ylabel('Largeur du sépale (cm)', fontsize=12)
plt.title('Dataset Iris: Longueur vs Largeur du Sépale', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print('Visualisation créée avec succès!')
```

### Étape 4 : Visualisation avancée - Pairplot

Le pairplot affiche les relations entre toutes les paires de variables. C'est un excellent outil pour explorer les données multidimensionnelles.

```python
# Créer un pairplot avec seaborn
pairplot = sns.pairplot(
    df_iris,
    hue='species',
    diag_kind='hist',
    plot_kws={'alpha': 0.7, 's': 80},
    height=2.5
)

pairplot.fig.suptitle('Pairplot du Dataset Iris', fontsize=16, fontweight='bold', y=1.001)
plt.tight_layout()
plt.show()

print('Pairplot généré avec succès!')
```

## Appel à une API Hugging Face

Hugging Face propose une API gratuite pour tester différents modèles de machine learning. Nous allons utiliser un modèle d'analyse de sentiments sur du texte en français.

### Option 1 : Utiliser l'API Hugging Face avec requests

Cette approche nécessite une clé API gratuite disponible sur https://huggingface.co/settings/tokens

```python
# Option 1 : Appel à l'API Hugging Face pour l'analyse de sentiments
import requests

# Vous devez remplacer par votre propre token Hugging Face
# Obtenez un token gratuit sur : https://huggingface.co/settings/tokens
HF_API_TOKEN = "hf_YOUR_TOKEN_HERE"  # À remplacer par votre token

# URL de l'API Hugging Face pour l'analyse de sentiments en français
api_url = "https://api-inference.huggingface.co/models/nlptown/bert-base-multilingual-uncased-sentiment"

headers = {"Authorization": f"Bearer {HF_API_TOKEN}"}

def analyser_sentiment(texte):
    """
    Fonction pour analyser le sentiment d'un texte via l'API Hugging Face.
    
    Args:
        texte (str): Le texte à analyser
    
    Returns:
        dict: Les résultats de l'analyse de sentiment
    """
    try:
        payload = {"inputs": texte}
        response = requests.post(api_url, headers=headers, json=payload)
        result = response.json()
        return result
    except Exception as e:
        return {"erreur": str(e)}

# Exemples de textes en français à analyser
textes_exemples = [
    "J'adore ce cours, c'est fantastique!",
    "Ce n'est pas très bon, j'en suis déçu.",
    "C'est acceptable mais pourrait être mieux."
]

print("Analyse de sentiments avec Hugging Face:")
print("="*60)

for texte in textes_exemples:
    print(f"\nTexte: {texte}")
    resultat = analyser_sentiment(texte)
    print(f"Résultat: {resultat}")
    print("-"*60)
```

### Option 2 : Utiliser la bibliothèque transformers localement

Cette approche utilise la bibliothèque `transformers` de Hugging Face pour exécuter le modèle localement, sans API.

```python
# Option 2 : Utiliser transformers pipeline localement (recommandé pour la démo)
from transformers import pipeline

# Charger le pipeline d'analyse de sentiments en français
# Cela télécharge le modèle la première fois
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)

print("Pipeline chargé avec succès!\n")

# Textes à analyser en français
textes_test = [
    "J'adore ce cours, c'est fantastique!",
    "Ce n'est pas très bon, j'en suis déçu.",
    "C'est acceptable mais pourrait être mieux.",
    "Le Machine Learning est passionnant et très utile!",
    "Je déteste attendre les résultats du training."
]

print("Analyse de sentiments avec transformers (exécution locale):")
print("="*70)

# Analyser chaque texte
for texte in textes_test:
    resultat = sentiment_pipeline(texte)[0]
    sentiment = resultat['label']
    confiance = resultat['score']
    
    print(f"\nTexte: {texte}")
    print(f"Sentiment: {sentiment}")
    print(f"Confiance: {confiance:.4f}")
    print("-"*70)
```
